```python
import os
import zipfile
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

from tensorflow.keras.preprocessing.image import ImageDataGenerator, load_img, img_to_array
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.utils import to_categorical
```

```python
extract_path = "split_dataset"

train_dir = os.path.join(extract_path, "train")
val_dir = os.path.join(extract_path, "val")

print("Train folder:", train_dir)
print("Validation folder:", val_dir)
```

```python
train_dir = os.path.join(extract_path, "train")
val_dir = os.path.join(extract_path, "val")

IMG_SIZE = 224
BATCH_SIZE = 16
EPOCHS = 20
```

```python
import os
import numpy as np

def collect_image_paths(base_dir):
    image_paths = []
    labels = []

    for cancer_type in os.listdir(base_dir):
        cancer_path = os.path.join(base_dir, cancer_type)

        if not os.path.isdir(cancer_path):
            continue

        # Handle normal class (no stages)
        if cancer_type.lower() == "normal":
            for img in os.listdir(cancer_path):
                img_path = os.path.join(cancer_path, img)
                if img.lower().endswith((".jpg", ".jpeg", ".png")):
                    image_paths.append(img_path)
                    labels.append("normal")
        else:
            # Go inside stage folders
            for stage in os.listdir(cancer_path):
                stage_path = os.path.join(cancer_path, stage)

                if not os.path.isdir(stage_path):
                    continue

                class_name = f"{cancer_type}/{stage}"

                for img in os.listdir(stage_path):
                    img_path = os.path.join(stage_path, img)
                    if img.lower().endswith((".jpg", ".jpeg", ".png")):
                        image_paths.append(img_path)
                        labels.append(class_name)

    return image_paths, labels


# Collect training and validation data
train_paths, train_labels = collect_image_paths(train_dir)
val_paths, val_labels = collect_image_paths(val_dir)

# Build class index
all_classes = sorted(set(train_labels))
class_indices = {cls: i for i, cls in enumerate(all_classes)}

# Print detected classes
print("Detected classes:")
for k, v in class_indices.items():
    print(v, "→", k)

print("\nTotal classes:", len(class_indices))

# Save class labels
np.save("class_labels.npy", class_indices)
```

```python
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.utils import to_categorical

def load_images(paths, labels, class_indices):
    X = []
    y = []

    for img_path, label in zip(paths, labels):
        img = load_img(img_path, target_size=(IMG_SIZE, IMG_SIZE))
        img_array = img_to_array(img) / 255.0
        X.append(img_array)
        y.append(class_indices[label])

    X = np.array(X)
    y = to_categorical(y, num_classes=len(class_indices))

    return X, y


# Load train and validation data
X_train, y_train = load_images(train_paths, train_labels, class_indices)
X_val, y_val = load_images(val_paths, val_labels, class_indices)

print("Train shape:", X_train.shape)
print("Validation shape:", X_val.shape)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.optimizers import Adam

model = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(IMG_SIZE, IMG_SIZE, 3)),
    MaxPooling2D(2,2),

    Conv2D(64, (3,3), activation='relu'),
    MaxPooling2D(2,2),

    Conv2D(128, (3,3), activation='relu'),
    MaxPooling2D(2,2),

    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(len(class_indices), activation='softmax')
])

model.compile(
    optimizer=Adam(learning_rate=0.0001),
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

model.summary()
print("\nTraining model...")
model.fit(
    X_train,
    y_train,
    validation_data=(X_val, y_val),
    epochs=EPOCHS,
    batch_size=BATCH_SIZE
)

model.save("cancer_model.h5")
print("\nModel saved as cancer_model.h5")
```

```python
def predict_image(image_path):

    model = load_model("cancer_model.h5")
    class_indices = np.load("class_labels.npy", allow_pickle=True).item()
    labels = {v: k for k, v in class_indices.items()}

    img = load_img(image_path, target_size=(IMG_SIZE, IMG_SIZE))
    img_array = img_to_array(img) / 255.0
    img_array = np.expand_dims(img_array, axis=0)
    prediction = model.predict(img_array)[0]

    sorted_indices = np.argsort(prediction)[::-1]
    top1_idx = sorted_indices[0]
    top2_idx = sorted_indices[1]

    top1_conf = prediction[top1_idx]
    top2_conf = prediction[top2_idx]

    top1_label = labels[top1_idx]
    top2_label = labels[top2_idx]
     # Strict Confidence
    if top1_conf < 0.75:
        print("❌ Rejected: Low confidence (Unknown Image)")
        return "Rejected", None, None

    # Allowed Types
    allowed_types = ["normal", "gynecological", "testicular"]

    def extract_type(label):
        if label.lower() == "normal":
            return "normal"
        return label.split("/")[0].lower()

    type1 = extract_type(top1_label)
    type2 = extract_type(top2_label)

    if type1 not in allowed_types:
        print("❌ Rejected: Unsupported organ")
        return "Rejected", None, None

    # Mixed Detection
    if type1 != type2 and top2_conf > 0.40:
        print("❌ Rejected: Mixed cancer types")
        return "Rejected", None, None
    # Stage Validation
    if type1 != "normal":
        parts = top1_label.split("/")
        stage = parts[1].lower()

        allowed_stages = ["stage1","stage2","stage3","stage4","critical"]

        if stage not in allowed_stages:
            print("❌ Rejected: Invalid stage")
            return "Rejected", None, None

        cancer_type = parts[0]
        cancer_stage = parts[1]
    else:
        cancer_type = "Normal"
        cancer_stage = "None"

    print("Cancer Type :", cancer_type)
    print("Cancer Stage:", cancer_stage)
    print("Confidence  :", round(top1_conf * 100, 2), "%")

    return cancer_type, cancer_stage, top1_conf
    
```

```python
from tensorflow.keras.models import load_model
import numpy as np
from tensorflow.keras.preprocessing.image import load_img, img_to_array
import os
```

```python
def predict_image(image_path):
    model = load_model("cancer_model.h5")
    class_indices = np.load("class_labels.npy", allow_pickle=True).item()

    labels = {v: k for k, v in class_indices.items()}

    img = load_img(image_path, target_size=(IMG_SIZE, IMG_SIZE))
    img_array = img_to_array(img) / 255.0
    img_array = np.expand_dims(img_array, axis=0)

    prediction = model.predict(img_array)
    predicted_class = np.argmax(prediction)
    confidence = np.max(prediction)

    label = labels[predicted_class]

    # Handle normal class
    if label.lower() == "normal":
        cancer_type = "Normal"
        cancer_stage = "None"
    else:
        # Split using "/" instead of os.sep
        parts = label.split("/")

        if len(parts) == 2:
            cancer_type = parts[0]
            cancer_stage = parts[1]
        else:
            cancer_type = label
            cancer_stage = "Unknown"

    print("Cancer Type :", cancer_type)
    print("Cancer Stage:", cancer_stage)
    print("Confidence  :", round(confidence * 100, 2), "%")
```
